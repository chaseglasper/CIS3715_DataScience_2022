# Part 2. Vector and Matrix

In this lab, we will learn how to use Python to operate vectors and matrices. 

## 1. Example

### 1.1 Load data with pandas

```python
import numpy as np
import pandas as pd

# load data with pandas
df = pd.read_csv('cars.csv')
# print(df)

# convert a pandas dataframe to a numpy array
X = df.values[:, 1:-1]
print(type(X))
print(X.shape)
print(X)
```

### 1.2 Operation for vector

```python
#Get first row and the second to last row

x1 = X[0, :]
x2 = X[-2, :]

print(x1)
print(x2)

# Addition

y = x1 + x2
print(y)
```

### 1.3 Operation for matrix

```python
# Get first 5 rows and last 5 rows 

m1 = X[0:5, :]
m2 = X[-5:, :]

print(m1)
print(m2)

# Subtraction
m = m1 - m2
print(m)
```

## 2 Task

### 2.1 Get the first column and the second to last column from X

```python
# this is your code to get two vectors

v1 = X[: , 0]
v2 = X[: , -2]
print(v1)
print(v2)
```

### 2.2 Perform following operations on these two vectors and print the result

```python
v1 = X[: , 0]
v2 = X[: , -2]

# 1. addition and subtraction

AddVector = v1 + v2

print(AddVector)
print("\n")

SubVector = v1 - v2

print(SubVector)
print("\n")

# 2. element-wise multiplication

MultVector = v1 * v2
print(MultVector)
print("\n")

# 3. inner product
Vector1 = np.array(v1)
Vector2 = np.array(v2)
print("inner product: {}".format(np.inner(Vector1, Vector2)))
print("\n")

# 4. three norms
print("Three Norms of Vector 1")
print("l1 norm: {}".format(np.linalg.norm(Vector1, ord=1)))
print("l2 norm: {}".format(np.linalg.norm(Vector1, ord=2)))
print("max-norm: {}".format(np.linalg.norm(Vector1, ord=np.inf)))
print("\n")


print("Three Norms of Vector 2 ")
print("l1 norm: {}".format(np.linalg.norm(Vector2, ord=1)))
print("l2 norm: {}".format(np.linalg.norm(Vector2, ord=2)))
print("max-norm: {}".format(np.linalg.norm(Vector2, ord=np.inf)))
print("\n")

# 5. compute the distance between these two vectors

print("distance: {}".format(np.linalg.norm(Vector1 - Vector2, ord=2)))
```

### 2.3 Get the first three columns and the last three columns from X

```python
# this is your code to get two matrices: A, B


v1 = X[: , 3]
v2 = X[: , -3]
Vector1 = np.array(v1)
Vector2 = np.array(v2)
print(Vector1)
print(Vector2)
```

### 2.4 Perform following operations on these two matrices and print the result

```python
# 1. addition and subtraction 

print("substraction: {}".format(v1-v2))
print("addition: {}".format(v1+v2))


# 2. compute the transpose of the second matrix: C = B^T, and print the shape of C

Vector3 = np.transpose(Vector2)
print("Transposition of B: {}".format(Vector3))
print("Shape of C : {}".format(Vector3.shape))

# 3. compute the product between A and C, and print the result and its shape

Vector4 = Vector1 * Vector3
print("Product of A and C: {}".format(Vector4))
print("Shape of Product: {}".format(Vector4.shape))
      
# 4. compute the transpose of the first matrix: D = A^T, and print the shape of D
Vector5 = np.transpose(Vector1)
print("Shape of D: {} ".format(Vector5.shape))

# 5. compute the product between D and B, and print the result and its shape
Vector6 = Vector5 * Vector2
print("Product of D and B: {}".format(Vector6))
print("Shape of Product: {}".format(Vector6.shape))
```

```python
list1 = ["apple", "banana", "cherry"]
list2 = [1, 5, 7, 9, 3]
list3 = [True, False, False]

print("Basic Lists\n")

print(list1)
print(list2[2])
print(list3[0:2])
print("\n")

list3.append(True)
list2.insert(2, 11)

list4 = ["mango", "pineapple", "kiwi"]
list1.extend(list4)

print("Using Append, Extend, and Insert with Lists\n")

print(list1)
print(list2)
print(list3)
print("\n")


print("using remove function to remove banana\n")
list1.remove("banana")
print(list1)
print("\n")


print("printing by looping through list\n")
for i in range(len(list1)):
    print(list1[i])
print("\n")


    
```

```python
print("Using tuples")

tuple1 = ("apple", "banana", "cherry")

print(tuple1)
print("\n")

print("extract values to variables")
(green, yellow, red) = tuple1
print(green + " " + yellow + " " + red + "\n")

print("Looping through a tuple")
for x in tuple1:
    print (x)




```

```python
print("Using Dictionaries")
dict1 = {
    "brand": "Ford" , 
    "model": "Mustang" , 
    "year": 1964   
}

print(dict1)
print(dict1["brand"])
print("\n")
print("accessing a dictionary")

x = dict1.keys()
print(x)

y = dict1.values()
print(y)

z = dict1.items()
print(z)
print("\n")
print("Adding items")

dict1["color"] = "red"
print(dict1)
print("\n")

print("Updating items")
dict1.update({"year" : 2022})
print(dict1)
print("\n")

print("Removing items")
dict1.pop("model")
print(dict1)
print("\n")

print("Looping through dictionary keys")
print("===keys===")
for x in dict1.keys():
    print(x)
    


```
